Getting company symbols Retuter

```
query = 'https://www.reuters.com/finance/stocks/lookup?search=Walmart&searchType=any&sortBy=&dateRange=&comSortBy=marketcap'

response = requests.get(query)

soup = BeautifulSoup(response.text, "html.parser")

table = soup.find('table', {'class': 'search-table-data'})
tr = table.findAll('tr')
list_data = []
for row in tr:
  string = str(row.findAll('td')[1:-1])
  string = string.replace('<td>','')
  string = string.replace('</td>','')
  string = string.replace(']','')
  string = string.replace('[','')
  list_data.append(string)
  
```

```
list_data = list_data[1:]
print(list_data)
```

```
stocks = list_data
df = pd.DataFrame(columns = ['Date' , 'Ticker' , 'News'])
c = 0
start_date = datetime.datetime(2019,1,1)
while(start_date <= end_date):
  date = start_date
  #file = open('/contents'  +  date.strftime('%Y-%m-%d') + '.csv', 'w')
  for i in range(len(stocks)):  
    query = 'https://www.reuters.com/finance/stocks/company-news/' + stocks[i] + '?date=' + format(date.month, '02d') + format(date.day, '02d') + str(date.year);
    print(query)
    response = requests.get(query)
    soup = BeautifulSoup(response.text, "html.parser")
    divs = soup.findAll('div', {'class': 'feature'})
    print('Found ' + str(len(divs)) + ' articles. For ' + stocks[i])

    for item in divs:
      df.loc[c] = [
                    str(date) ,
                    str(stocks[i]),
                    str(item.find("h2").find("a").string)
                  ]
      c += 1
  start_date += datetime.timedelta(days = 1) 
```
